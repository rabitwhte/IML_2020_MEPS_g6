```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("MEPS_data_preprocessed.csv")
data.drop(columns = ["PANEL", "PERSONWT"], inplace = True)
```

```python
data.shape
```

```python
0.3 * data.shape[0]
```

```python
data.columns
```

```python
def plots(data, col_name):
    plt.figure(figsize = (10, 4))
    data[[col_name]].boxplot(vert = False)
    plt.show()
    data[[col_name]].hist()
    plt.show()
    f = plt.figure(figsize = (10, 10))
    plt.matshow(data.corr(), fignum = f.number)
    plt.show()
```

```python
plots(data, "HEALTHEXP")
```

```python
train, test = train_test_split(data, test_size = 0.3, random_state = 123)
```

```python
plots(train, "HEALTHEXP")
```

```python
plots(test, "HEALTHEXP")
```

```python
train.to_csv("MEPS_data_preprocessed_train.csv", index = False)
test.to_csv("MEPS_data_preprocessed_test.csv", index = False)
```
